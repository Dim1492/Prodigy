```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(r'C:\Users\VICTUS\Downloads\age_gender.csv\age_gender.csv')
                 





```

```python
# Histogram for continuous variable (age)
plt.figure(figsize=(8, 6))
sns.histplot(df['age'], bins=10, kde=True)  # Adjust bins as needed
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar chart for categorical variable (ethnicity)
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', data=df)
plt.title('Distribution of gender')
plt.xlabel('gender')
plt.ylabel('Count')
plt.show()
```
